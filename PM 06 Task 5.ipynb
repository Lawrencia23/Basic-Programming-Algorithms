```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
iris = fetch_ucirepo(id=53) 
  
# data (as pandas dataframes) 
X = iris.data.features 
y = iris.data.targets 
  
# metadata 
print(iris.metadata) 
  
# variable information 
print(iris.variables) 
```

```python
import pandas as pd

df = pd.concat([X, y], axis=1)
df.head()

```

```python
df = df.rename(columns={"class": "species"})
```

```python
mean_val = df['sepal length'].mean()
median_val = df['sepal length'].median()
variance_val = df['sepal length'].var()

print("Mean:", mean_val)
print("Median:", median_val)
print("Variance:", variance_val)
```

```python
df['petal_length_normalized'] = (
    (df['petal length'] - df['petal length'].min()) /
    (df['petal length'].max() - df['petal length'].min())
)

df[['petal length', 'petal_length_normalized']].head()
```

```python
def linear_search_species(df, target):
    results = []
    for i in range(len(df)):
        if df.iloc[i]['species'] == target:
            results.append(df.iloc[i])
    return results

setosa_rows = linear_search_species(df, "Iris-setosa")
print("Rows found:", len(setosa_rows))
```

```python
df.query("species == 'Iris-setosa'")
```

```python
def bubble_sort_by_sepal(df):
    data = df.to_dict('records')   # convert rows to list of dicts
    n = len(data)

    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j]['sepal length'] > data[j + 1]['sepal length']:
                data[j], data[j + 1] = data[j + 1], data[j]

    return pd.DataFrame(data)

sorted_bubble = bubble_sort_by_sepal(df)
sorted_bubble.head()
```
